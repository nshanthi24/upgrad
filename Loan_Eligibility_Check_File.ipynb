### Find out nulls 

```python
print(loan_df.isnull().sum())
```

```python
# Find out columns having Null values more than 50% data
obj = loan_df.isnull().sum()
list_nulls = []

for key,value in obj.iteritems():
    if value/39717 > 0.5:
        list_nulls.append(key)
        #print(key,'- ',value/39717)
print(list_nulls)
```

```python
# list of all columns
total_list = loan_df.columns
```

```python
non_null_list = list(set(total_list) - set(list_nulls))
print(len(list_nulls),len(non_null_list),len(total_list))
```

```python
new_loan_df = loan_df.drop(list_nulls, axis = 1)
new_loan_df.shape
```

```python
new_loan_df.isnull().sum()
```

```python
# We can remove columns which resemble post loan approval features
new_loan_df=new_loan_df.drop(['delinq_2yrs','earliest_cr_line', 'open_acc','inq_last_6mths','pub_rec','revol_bal', 'revol_util',
                 'total_acc','out_prncp', 'out_prncp_inv', 'total_pymnt', 'total_pymnt_inv','total_rec_prncp',
                  'total_rec_late_fee','recoveries','collection_recovery_fee','last_pymnt_d','last_credit_pull_d',
                  'application_type'
                 ],axis=1)
```

```python
new_loan_df.columns
```

```python
# id columns can be removed as well and desc too
new_loan_df = new_loan_df.drop(['id','member_id','desc','collections_12_mths_ex_med','chargeoff_within_12_mths',
                                'initial_list_status'],axis=1)
```

```python
new_loan_df.head(6)
```
